# 01 Importando dataset 

```python
from sklearn import datasets
```

```python
import sklearn
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn import linear_model, metrics
import numpy as np
import pandas as pd
```

```python
X, y = datasets.load_diabetes(return_X_y=True)
```

```python
X_ = pd.DataFrame(X)
y_ = pd.DataFrame(y)
```

# 02 Gráficos de correlação. 

```python
count = 0
fig, axes = plt.subplots(4, 3, sharex=True, sharey=True)
for i in range(4):
    for j in range(3):
        if i == 3 and j == 1:
            break
        axes[i, j].scatter(X_[count],y_)
        count += 1
plt.subplots_adjust(left=7, bottom=7, right=15, top=15,
wspace=None, hspace=None)
```




<h2> 06. Predição para todos os dados de entrada </h2>





```python
y_pred = []
y_pred = model.predict(X_)
```

<h2> 07. Avalie o modelo através da métrica MSE. </h2>

<h3> Scikit Learn </h3>

```python
mse = metrics.mean_squared_error(y, y_pred)
print(mse)
```

<h3> Somente o Python ("Manualmente") </h3>

```python
mse_braco = sum((y_.values - y_pred) ** 2) / len(y_)
print(mse_braco)
```

<h2> 08. Modelo de regressão linear para a feature que melhor se ajusta a uma função linear. </h2>

```python
modelplayboy = linear_model.LinearRegression()
```

```python
Xlinha = [[i] for i in X_[2]]
modelplayboy.fit(Xlinha, y_)
```

<h3> A) </h3>
<h4> b1 </h4>

```python
b1 = np.sum((Xlinha - np.mean(Xlinha)) * (y_ - np.mean(y_))) / np.sum((Xlinha - np.mean(Xlinha)) ** 2) 
b1
```

<h4> b0</h4>

```python
b0 = np.mean(y_) - b1 * np.mean(Xlinha)
b0
```

<h3> B) </h3>
<h4> b1 </h4>

```python
b12 = modelplayboy.coef_
b12
```

<h4> b0 </h4>

```python
b02 = model.intercept_
b02
```

<h3> C) </h3>

```python
y_pred_ = []
y_pred_ = modelplayboy.predict(Xlinha)
```

```python
plt.grid()
plt.scatter(Xlinha, y_, color='orange')
plt.plot(Xlinha, y_pred_)
```

<h3> D)</h3>

```python
mse = metrics.mean_squared_error(y, y_pred_)
print(mse)
```
